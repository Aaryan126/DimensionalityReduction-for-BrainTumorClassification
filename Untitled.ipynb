```python
import os
import numpy as np
import cv2  # OpenCV for image processing
from tqdm import tqdm  # Import tqdm for progress bar
import matplotlib.pyplot as plt

# Paths to the dataset
train_dir = 'Brain_Tumor_Dataset/brain/train'
#val_dir = 'Brain_Tumor_Dataset/brain/val'
test_dir = 'Brain_Tumor_Dataset/brain/test'

# Function to load images and labels
def load_images_and_labels(directory):
    images = []
    labels = []

    # Get total number of images for the progress bar
    total_images = sum(len(files) for _, _, files in os.walk(directory))
    
    # Use a single tqdm progress bar for all images
    with tqdm(total=total_images, desc=f"Loading images from {directory}") as pbar:
        for label in os.listdir(directory):
            label_path = os.path.join(directory, label)
            if os.path.isdir(label_path):
                for img_file in os.listdir(label_path):
                    img_path = os.path.join(label_path, img_file)
                    img = cv2.imread(img_path)
                    if img is not None:
                        img = cv2.resize(img, (128, 128))
                        #img = cv2.resize(img, (256, 256))  # Resize if needed
                        #img = cv2.resize(img, (128, 128))  # Resize if needed
                        images.append(img)
                        labels.append(label)
                    pbar.update(1)  # Update progress bar for each image loaded
    
    return np.array(images), np.array(labels)

# Load the datasets
X_train, y_train = load_images_and_labels(train_dir)
#X_val, y_val = load_images_and_labels(val_dir)
X_test, y_test = load_images_and_labels(test_dir)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
import numpy as np

# Move the last sample from test to train
X_train = np.append(X_train, X_test[-1:,:,:,:], axis=0)
y_train = np.append(y_train, y_test[-1:], axis=0)

# Remove the last sample from test
X_test = X_test[:-1]
y_test = y_test[:-1]

print("New shapes:")
print("X_train:", X_train.shape)
print("y_train:", y_train.shape)
print("X_test:", X_test.shape)
print("y_test:", y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

X_train_flat = X_train.reshape(X_train.shape[0], -1)
X_train_flat = X_train_flat.astype(np.float32)
X_test_flat = X_test.reshape(X_test.shape[0], -1)
X_test_flat = X_test_flat.astype(np.float32)

scaler = StandardScaler()

X_train_flat = scaler.fit_transform(X_train_flat)
X_test_flat = scaler.fit_transform(X_test_flat)
```

```python
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt

# Initialize PCA
pca = PCA()
# Fit PCA to the flattened training data
pca.fit(X_train_flat)

# Get the explained variance ratio
explained_variance = pca.explained_variance_ratio_

# Calculate cumulative explained variance
cumulative_variance = np.cumsum(explained_variance)

# Plot explained variance
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance) + 1), cumulative_variance, marker='o', linestyle='--')
plt.title('Cumulative Explained Variance by Principal Components')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.axhline(y=0.90, color='r', linestyle='--')  # Line for 90% explained variance
plt.axvline(x=np.argmax(cumulative_variance >= 0.90) + 1, color='g', linestyle='--')  # Number of components for 90%
plt.grid()
plt.show()

# Print the number of components to achieve 90% variance
num_components_90 = np.argmax(cumulative_variance >= 0.90) + 1
print(f"Number of components to retain at least 90% variance: {num_components_90}")
num_components_95 = np.argmax(cumulative_variance >= 0.95) + 1
print(f"Number of components to retain at least 95% variance: {num_components_95}")
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=629)
pca.fit(X_train_flat, y_train)
```

```python
X_train_pca = pca.transform(X_train_flat)
X_test_pca = pca.transform(X_test_flat)
```

```python
import numpy as np
from sklearn.model_selection import KFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Assuming you already have the following variables:
# X_train_pca: PCA-transformed training features
# X_test_pca: PCA-transformed test features
# y_train: labels for the training set
# y_test: labels for the test set

# Set up k-NN classifier and parameters
k_values = range(1, 41, 2)  # Test even values from 2 to 40
best_k = None
best_test_accuracy = 0

# Evaluate each k
for k in k_values:
    # Create and fit the k-NN model on the entire training set
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_pca, y_train)

    # Make predictions on the test set
    y_test_pred = knn.predict(X_test_pca)

    # Calculate test accuracy
    test_accuracy = accuracy_score(y_test, y_test_pred)

    # Print the test accuracy for the current k
    print(f"k={k}: Test Accuracy = {test_accuracy:.4f}")

    # Update the best k if needed
    if test_accuracy > best_test_accuracy:
        best_test_accuracy = test_accuracy
        best_k = k

print(f"\nBest k: {best_k} with test accuracy: {best_test_accuracy:.4f}")
```

```python
from sklearn.svm import SVC, SVR
from sklearn.linear_model import SGDClassifier, RidgeClassifier, ElasticNet, Perceptron
from sklearn.metrics import classification_report
import time

# Define a list of classifiers to test
classifiers = {
    "KNN": KNeighborsClassifier(n_neighbors=1, n_jobs=1),
    "Support Vector Classifier": SVC(kernel='linear', max_iter=10000),
    "Stochastic Gradient Descent Classifier": SGDClassifier(max_iter=5000, tol=1e-3, random_state=42),
    "Ridge Classifier": RidgeClassifier(),
    "Perceptron": Perceptron(max_iter=1000, tol=1e-3)
}

# Placeholder for results
results = {}

# Loop through each classifier
for name, clf in classifiers.items():
    print(f"Training {name}...")
    
    # Start the timer
    start_time = time.time()
    
    # Train the model
    clf.fit(X_train_flat, y_train)
    
    # Stop the timer after training completes
    end_time = time.time()
    
    # Calculate the duration
    training_duration = end_time - start_time
    print(f"{name} training completed in {training_duration:.2f} seconds.")
    
    # Make predictions on the test set
    y_pred = clf.predict(X_test_flat)
    
    # Evaluate the model
    report = classification_report(y_test, y_pred, target_names=clf.classes_, digits=4)
    results[name] = report
    print(report)

# Print summary of results
print("Summary of Results:")
for name, report in results.items():
    print(f"\n{name} Classification Report:\n{report}")
```
